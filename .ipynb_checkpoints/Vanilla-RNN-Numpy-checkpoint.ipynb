```python
# only one import numpy, Can't help but feel a little hardcore doing that
import numpy as np
```

```python
# Data import, Using same Tiny Shakespeare dataset
data = open('tinyshakespeare.txt', 'r').read() #see my other generate txt and language model about scraping docx flies from a directory
print(data[0:250])
```

```python
chars = list(set(data))
print(chars)   # chars is a list of all characters found in the document
```

```python
data_size, vocab_size = len(data), len(chars)
print('data_size = {}, vocab_size = {}'.format(data_size, vocab_size))
```

```python
char_to_ix = {ch:i for i,ch in enumerate(chars)}  # Ibelieve this is a dictionary comprehension... only done list comprehensions
print(char_to_ix) # provides every character a value... characters are the keys
```

```python
ix_to_char = {i:ch for i,ch in enumerate(chars)}
print(ix_to_char) # reverse of above
```

```python
# Hyperparameters

hidden_size = 100  # size of the hiddn layer
seq_length = 25 # The length of charectures to train from I think... Will check below
learning_rate = 1e-1 # wonder why the learning rate is relativly large... will explor
```

```python
# model parameters
# terms used below W = weight, b = bias, x = input, h = hidden, y = output:  e.g. Wxh = input weights to hidden
Wxh = np.random.randn(hidden_size, vocab_size)*0.01 # input to hidden, Initialized with small weights, *0.01 is to make values very small
Whh = np.random.randn(hidden_size, hidden_size)*0.01 # hidden to hidden
Why = np.random.randn(vocab_size, hidden_size)*0.01 # hidden to output
bh = np.zeros((hidden_size, 1)) # hidden bias
by = np.zeros((vocab_size, 1)) # output bias

print('Wxh shape = {}'.format(Wxh.shape))
print('Wxh = {}'.format(Wxh))
print('bh shape = {}'.format(bh.shape))
print('bh = {}'.format(bh))
```

```python
# loss function
def lossFun(inputs, targets, hprev):
    """
    inputs, targets are both lists of intergers (see below).
    hprev is an array fo size (hidden x 1) represents hidden memory and is reset to a zero at every epoch
    returns the loss, gradient on model parameters and the last hidden state
    """
    xs, hs, ys, ps = {}, {}, {}, {} # inputs, hidden, outputs, probabilities
    hs[-1] = np.copy(hprev) # 
    loss = 0 # initilize to 0
    # forward pass
    for t in range(len(inputs)): # xrange is covered by range in python 3
        xs[t] = np.zeros((vocab_size,1)) # basically encoding in a 1 by 65 array
        xs[t][inputs[t]] = 1
        hs[t] = np.tanh(np.dot(Wxh, xs[t]) + np.dot(Whh, hs[t-1]) + bh) #pushing input weghts through tanh function
        ys[t] = np.dot(Why, hs[t]) + by # unnormalized log probabilities for next chars
        ps[t] = np.exp(ys[t]) / np.sum(np.exp(ys[t])) # probabilities for next char
        loss += -np.log(ps[t][targets[t],0]) #Karpathy notes this as the softmax but I believe the above is the softmax this is just the negative log loss based on comparing the prediction to the targets
    # Backward pass: calculate those gradients
    dWxh, dWhh, dWhy = np.zeros_like(Wxh), np.zeros_like(Whh), np.zeros_like(Why) # backprop of weight changes creating new arrays of same shape
    dbh, dby = np.zeros_like(bh), np.zeros_like(by)
    dhnext = np.zeros_like(hs[0])
    for t in reversed(range(len(inputs))):
        dy = np.copy(ps[t])
        #print('dy = {}'.format(dy))
        dy[targets[t]] -= 1 # backprop into y. see http://cs231n.github.io/neural-networks-case-study/#grad if confused here
        #print('dy[targets[t]] = {}'.format(dy[targets[t]]))
        dWhy += np.dot(dy, hs[t].T) # Transpose of hs[t]
        #print('dWhy = {}'.format(dWhy))
        dby += dy
        #print('dby = {}'.format(dby))
        dh = np.dot(Why.T, dy) + dhnext # backprop into h
        #print('dh = {}'.format(dh))
        dhraw = (1 - hs[t] * hs[t]) * dh # backprop through tanh nonlinearity
        #print('dhraw = {}'.format(dhraw))
        dbh += dhraw
        dWxh += np.dot(dhraw, xs[t].T)
        dWhh += np.dot(dhraw, hs[t-1].T)
        dhnext = np.dot(Whh.T, dhraw)
    for dparam in [dWxh, dWhh, dWhy, dbh, dby]:
        np.clip(dparam, -5, 5, out=dparam) # clip to mitigate exploding gradients
    return loss, dWxh, dWhh, dWhy, dbh, dby, hs[len(inputs)-1]
```

```python
def sample(h, seed_ix, n):
    """ 
    sample a sequence of integers from the model 
    h is memory state, seed_ix is seed letter for first time step
    """
    x = np.zeros((vocab_size, 1))
    x[seed_ix] = 1 #like a one hot encoding of the seed char
    ixes = []
    for t in range(n):
        h = np.tanh(np.dot(Wxh, x) + np.dot(Whh, h) + bh)
        y = np.dot(Why, h) + by
        p = np.exp(y) / np.sum(np.exp(y))
        ix = np.random.choice(range(vocab_size), p=p.ravel())
        x = np.zeros((vocab_size, 1))
        x[ix] = 1 # new one hot encoding of predicted char 
        ixes.append(ix) # append char number to list so it can be converted to letters after
    return ixes
```

```python
n, p = 0, 0
mWxh, mWhh, mWhy = np.zeros_like(Wxh), np.zeros_like(Whh), np.zeros_like(Why)
mbh, mby = np.zeros_like(bh), np.zeros_like(by) # memory variables for Adagrad filled with zeros
smooth_loss = -np.log(1.0/vocab_size)*seq_length # loss at iteration 0, this is the initial loss, like 2.3 is for a 10 option softmax here it is 104.3... as confirmed by first loss
while True:
    # prepare inputs (we're sweeping from left to right in steps seq_length long)
    if p+seq_length+1 >= len(data) or n == 0:
        hprev = np.zeros((hidden_size,1)) # reset RNN memory
        p = 0 # go from start of data
    inputs = [char_to_ix[ch] for ch in data[p:p+seq_length]] # important input is like "lets go to the stor"
    targets = [char_to_ix[ch] for ch in data[p+1:p+seq_length+1]] # output is like "ets go to the store"
    
    # The above is important because when people talk about char-rnn they often assume you are taking a single letter and predicting the next out of any context
    # unsuprisingly when i changed the model to take a seqence of 2 charecters using the first to predict the second, it did not learn any words

    # sample from the model now and then
    if n % 100 == 0: #every 100 iterations print sample string
        sample_ix = sample(hprev, inputs[0], 200) # previous hidden state, random seed, 200 char length
        txt = ''.join(ix_to_char[ix] for ix in sample_ix) # create string, turn numbers back in the char they represent
        print('----\n %s \n----' % (txt, )) 

    # forward seq_length characters through the net and fetch gradient
    loss, dWxh, dWhh, dWhy, dbh, dby, hprev = lossFun(inputs, targets, hprev)
    smooth_loss = smooth_loss * 0.999 + loss * 0.001
    if n % 100 == 0: print('iter %d, loss: %f' % (n, smooth_loss)) # print progress
  
    # perform parameter update with Adagrad
    # fancier update that just the typical gradient decent update
    for param, dparam, mem in zip([Wxh, Whh, Why, bh, by], 
                                [dWxh, dWhh, dWhy, dbh, dby], 
                                [mWxh, mWhh, mWhy, mbh, mby]):
        mem += dparam * dparam
        param += -learning_rate * dparam / np.sqrt(mem + 1e-8) # adagrad update

    p += seq_length # move data pointer, this just goes over the whole corpus 25 char at a time. wonder if random samples would make a diff
    n += 1 # iteration counter 
```
